```python
import cv2
import numpy as np

# Path to your screenshot
image_path = r"C:\Users\Sambhavi Roy\Downloads\4cycle\part 2.png"  # <-- Change to your screenshot name

# Load original image
original = cv2.imread(image_path)
if original is None:
    raise FileNotFoundError(f"Could not load image at {image_path}")

# Resize for display only
scale_percent = 50  # Display at 50% size (change if needed)
width = int(original.shape[1] * scale_percent / 100)
height = int(original.shape[0] * scale_percent / 100)
resized = cv2.resize(original, (width, height))

clone = resized.copy()
roi_points_display = []
roi_points_original = []

def draw_polygon(event, x, y, flags, param):
    global roi_points_display, roi_points_original, resized, original

    if event == cv2.EVENT_LBUTTONDOWN:
        roi_points_display.append((x, y))
        # Map display point to original coordinates
        orig_x = int(x / scale_percent * 100)
        orig_y = int(y / scale_percent * 100)
        roi_points_original.append((orig_x, orig_y))

        cv2.circle(resized, (x, y), 4, (0, 255, 0), -1)
        if len(roi_points_display) > 1:
            cv2.line(resized, roi_points_display[-2], roi_points_display[-1], (255, 0, 0), 2)
        cv2.imshow("Select ROI - ESC to finish", resized)

print("\nüñ±Ô∏è Click to select ROI points (clockwise). Press ESC to finish.\n")

cv2.imshow("Select ROI - ESC to finish", resized)
cv2.setMouseCallback("Select ROI - ESC to finish", draw_polygon)

while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 27:  # ESC
        if len(roi_points_display) > 2:
            cv2.line(resized, roi_points_display[-1], roi_points_display[0], (255, 0, 0), 2)
        break

cv2.destroyAllWindows()

# Output original coordinates
print("\nüìç Your ROI polygon coordinates (original scale):")
print("road_polygon_ref = np.array([")
for point in roi_points_original:
    print(f"    [{point[0]}, {point[1]}],")
print("], np.int32)")
```
