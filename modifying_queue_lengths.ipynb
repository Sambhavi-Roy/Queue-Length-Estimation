```python
import pandas as pd

# Load the CSV
df = pd.read_csv('/content/queue_data.csv')  # Replace with your actual file name

# Frame indices
start_decreasing = 1590
start_zero = 1920

# Make everything from 1:04 (frame 1920) onward zero
df.loc[df['Frame'] >= start_zero, 'Queue_Length_m'] = 0

# Ensure decreasing or constant from 0:53 (1590) to 1:03 (1919)
for i in range(start_decreasing + 1, start_zero):
    prev = df.loc[i - 1, 'Queue_Length_m']
    current = df.loc[i, 'Queue_Length_m']
    if current > prev:
        df.loc[i, 'Queue_Length_m'] = prev

# Ensure increasing or constant from start to 0:52 (frame 0 to 1589)
for i in range(1, start_decreasing):
    prev = df.loc[i - 1, 'Queue_Length_m']
    current = df.loc[i, 'Queue_Length_m']
    if current < prev:
        df.loc[i, 'Queue_Length_m'] = prev

# Save the updated CSV
df.to_csv('queue_corrected.csv', index=False)
print("Saved corrected CSV as 'queue_corrected.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV
df = pd.read_csv("/content/queue_corrected.csv")  # Replace with your actual CSV path

# Assuming 30 FPS video
fps = 30

# Add a 'Second' column
df['Second'] = (df['Frame'] // fps) + 1

# Compute average queue length for each second
avg_queue_per_sec = df.groupby('Second')['Queue_Length_m'].mean().reset_index()

# Save to CSV
avg_queue_per_sec.to_csv("avg_queue_per_second.csv", index=False)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(avg_queue_per_sec['Second'], avg_queue_per_sec['Queue_Length_m'], marker='o')
plt.xlabel('Time (seconds)')
plt.ylabel('Average Queue Length (m)')
plt.title('Average Queue Length Per Second')
plt.grid(True)
plt.tight_layout()
plt.show()
```
